```python
import pandas as pd
import numpy as np
import os
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Definir la ruta y los nombres de las columnas (este dataset no tiene encabezado)
ruta_adult = '/Users/mac/Desktop/datasets/adult/adult/adult.data' 
columnas = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]

try:
    # 1. Cargar el archivo original
    df_crudo = pd.read_csv(ruta_adult, names=columnas, sep=',', skipinitialspace=True)
    print(f"Dataset original cargado: {df_crudo.shape}")

    # 2. ELIMINACIÓN DE CARACTERÍSTICAS (Aquí es donde lo pones)
    # Por ejemplo, eliminamos 'fnlwgt' y 'education'
    df = df_crudo.drop(columns=['fnlwgt', 'education']) 
    
    # 3. Imprimir dimensiones finales
    print(f"✅ Dimensiones después de eliminar columnas: {df.shape}")
    
    # Mostrar las primeras filas para verificar
    display(df.head())

except Exception as e:
    print(f" Error: {e}")
```

```python
# 1. Convertir el objetivo (income) a números (0 y 1)
df_crudo['income'] = df_crudo['income'].apply(lambda x: 1 if '>50K' in str(x) else 0)

# --- IMPRESIÓN INICIAL ---
print(f"Dataset Original: {df_crudo.shape[0]} filas y {df_crudo.shape[1]} columnas")

# 2. Selección de variables numéricas y eliminación de características
# Aquí es donde se filtran las columnas
df_num = df_crudo.select_dtypes(include=[np.number])

# 3. Rellenar nulos si existen
df_num = df_num.fillna(df_num.median())

# 4. Separar X (características) e Y (objetivo)
X_raw = df_num.drop('income', axis=1).values
y_raw = df_num['income'].values.reshape(-1, 1)

# --- IMPRESIÓN DESPUÉS DE SELECCIÓN ---
print(f"Variables seleccionadas para el modelo: {df_num.columns.tolist()}")
print(f"Dimensiones de X (características): {X_raw.shape}")
print(f"Dimensiones de Y (objetivo): {y_raw.shape}")

# 5. Normalización
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X_raw)

# 6. División de datos
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_raw, test_size=0.2, random_state=42)

# --- IMPRESIÓN FINAL DE PREPARACIÓN ---
print("-" * 50)
print("Preparación completada. Datos listos para PyTorch.")
print(f"Conjunto de Entrenamiento: {X_train.shape[0]} filas")
print(f"Conjunto de Prueba: {X_test.shape[0]} filas")
print("-" * 50)
```
