El dataset sirve para predecir cuántas bicicletas se van a alquilar en una hora específica basándose en el clima y el tiempo (hora, mes, día).

Por qué se eliminan esas 4 columnas:

instant y dteday (Ruido): Son solo un número de fila y una fecha en texto. A la IA no le sirven porque los datos temporales útiles ya están en otras columnas numéricas (año, mes, hora).

casual y registered (Las Tramposas): Tu meta es predecir el total de alquileres (cnt). Como Total = Casual + Registrados, si dejas estas columnas, la IA simplemente aprenderá a sumarlas y hará trampa. Cuando intentes predecir el futuro real (donde no conoces a los casuales ni registrados), el modelo fallará. Esto se llama Fuga de Datos.

```python
# FASE 2.1: ELIMINACIÓN DE COLUMNAS (REFORZADA)
# 1. Limpiamos posibles espacios en blanco en los nombres de las columnas
df_crudo.columns = df_crudo.columns.str.strip()

# 2. Definimos las columnas a eliminar
cols_a_borrar = ['instant', 'dteday', 'casual', 'registered']

# 3. Borramos solo si existen para evitar el KeyError
df = df_crudo.drop(columns=[c for c in cols_a_borrar if c in df_crudo.columns])

print(" FASE 2.1: Limpieza completada.")
print(f"Columnas actuales: {list(df.columns)}")
print(f"Dimensiones finales: {df.shape}")
```
