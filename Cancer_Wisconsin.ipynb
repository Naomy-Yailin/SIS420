
El conjunto de datos original Wisconsin Breast Cancer contiene 11 columnas que describen biopsias de tumores para enseñar a una Inteligencia Artificial a diagnosticar el cáncer de mama, donde tu meta final es predecir la variable objetivo `class`(que vale 2 si el tumor es benigno y 4 si es maligno). Para lograr esta predicción, el modelo analiza 9 características biológicas de las células, todas evaluadas en una escala médica del 1 al 10: el grosor de la masa (si se agrupan en capas anormales), la uniformidad de su tamaño y forma(las células cancerosas suelen ser muy irregulares), la adhesión marginal(tendencia a separarse para invadir otros tejidos), el tamaño de la célula epitelial individual** (si están anormalmente agrandadas), la presencia de núcleos desnudos (núcleos que han perdido su líquido), la textura de la cromatina blanda (material genético con apariencia gruesa o manchada), la cantidad de nucléolos normales y la tasa de multiplicación o mitosis(velocidad a la que se dividen).  Se elimina la columna `id` (el número de expediente o registro del paciente). Es obligatorio quitar este dato porque no tiene valor médico; si se lo deja, El modelo se confundirá y correrá el grave riesgo de intentar memorizar los números de los pacientes para adivinar el resultado, en lugar de aprender a detectar los verdaderos patrones celulares que definen la enfermedad.
