###  Descripción del Dataset
Este conjunto de datos contiene solicitudes de tarjetas de crédito. Para proteger la confidencialidad, todos los nombres de los atributos y sus valores han sido reemplazados por símbolos sin significado aparente (A1 al A15).

Es un dataset clásico en Machine Learning (UCI Repository) debido a su excelente mezcla de variables continuas y categóricas, ideal para probar modelos de **Clasificación Binaria**.

###  Estructura de Datos
* **Instancias Totales:** 690 solicitudes.
* **Atributos Predictores (X):** 14 (A1 - A14).
    * **6 Numéricos:** Variables continuas (como ingresos o deudas).
    * **8 Categóricos:** Variables nominales (como estado civil o tipo de empleo).
* **Variable Objetivo (y):** `A15` (Clase de aprobación).
    * **Clase 1 (-):** Crédito Denegado (55.5%).
    * **Clase 2 (+):** Crédito Aprobado (44.5%).

```python
# FASE 1: IMPORTACIONES
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from IPython.display import display

%matplotlib inline
print("✅ Fase 1: Entorno listo para el análisis de crédito.")
```

```python
# FASE 2: CARGA DE DATOS
ruta = '/content/sample_data/australian.dat'

# Nombres de columnas según la descripción técnica
nombres_cols = [f'A{i}' for i in range(1, 16)]

# Cargamos el archivo (suele venir separado por espacios)
try:
    df_crudo = pd.read_csv(ruta, sep='\s+', header=None, names=nombres_cols)
    print(f"✅ Archivo '{ruta}' cargado con éxito.")
except Exception as e:
    print(f"❌ Error al cargar: {e}")

print(f"Dimensiones: {df_crudo.shape[0]} aplicaciones de crédito x {df_crudo.shape[1]} columnas.")
display(df_crudo.head(10))
```

```python
# FASE 2.1: AUDITORÍA DE DATOS
print("🔍 REPORTE TÉCNICO DE COLUMNAS:")
# Información de tipos y nulos
info_df = pd.DataFrame({
    'Tipo': df_crudo.dtypes,
    'Nulos': df_crudo.isnull().sum(),
    'Valores Únicos': df_crudo.nunique()
})
display(info_df)

# Separación técnica según el documento:
numericas = ['A2', 'A3', 'A7', 'A10', 'A13', 'A14']
categoricas = ['A1', 'A4', 'A5', 'A6', 'A8', 'A9', 'A11', 'A12']
target = 'A15'

print(f"\n📊 RESUMEN:")
print(f"- Atributos Numéricos: {len(numericas)}")
print(f"- Atributos Categóricos: {len(categoricas)}")
print(f"- Variable Objetivo: {target} (1: -, 2: +)")
```

```python
# FASE 2.3: SEPARACIÓN DE MATRICES (PREDICTORES VS TARGET)

# 1. Definimos X (Todas las columnas excepto A15)
X_df = df_crudo.drop(columns=['A15'])

# 2. Definimos y (Solo la columna A15)
y_df = df_crudo['A15']

print("🏗️ ARQUITECTURA DE DATOS:")
print("-" * 30)

# Muestra de X
print("\n📋 MATRIZ X (Predictores del Crédito - Primeras 5 filas):")
from IPython.display import display
display(X_df.head(5))

print("-" * 30)

# Muestra de y
print("\n🎯 VECTOR y (Target A15: 1=Denegado, 2=Aprobado - Primeras 5 filas):")
display(y_df.head(5))

print("-" * 30)
print(f"✅ Separación lista: X tiene {X_df.shape[1]} atributos, y es una sola columna de etiquetas.")
```
